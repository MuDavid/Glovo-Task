```python
import matplotlib.pyplot as plt

# Promotion period weeks
weeks_promo = [5, 6, 7, 8]

# Performance marketing costs for promo weeks
performance_marketing_costs = [3000, 3250, 3500, 3750]

# Billboard cost per week
billboard_cost_per_week = 500 * 10

# Calculate total costs per week
total_costs_per_week = [pm + billboard_cost_per_week for pm in performance_marketing_costs]

# Plotting the bar chart with values at the top
plt.figure(figsize=(10, 6))
bars = plt.bar(weeks_promo, total_costs_per_week, color='blue', alpha=0.6, label='Total Campaign Costs')

# Adding the values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # va is  vertical alignment

plt.title('Glovo Campaign Costs During Promotion Weeks')
plt.xlabel('Week')
plt.ylabel('Cost (€)')
plt.legend()
plt.grid(False)
plt.show()
```
